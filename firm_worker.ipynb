# Повторение переговоров о зарплате работника и фирмы

```python
import pandas as pd
import numpy as np
```

```python
pd.read_excel('firm_and_worker_2019-10-25.xlsx', header=None, sheet_name=1)
```

- Бот выбирает 65 за фирму и соглашается на 40 и выше за работника.
- Есть параметр, который определяет сорвался проект или нет, если усилия работника были равны 0. (group.fail > 25 - проект срывается)

```python
data = pd.read_excel('firm_and_worker_2019-10-25.xlsx', sheet_name=0)
```

```python
data['group.fail'] = data['group.fail'].apply(lambda x: int(x > 25))
```

```python
data
```

**Рассмотрим первую сессию.**

```python
data[['participant.label', 'group.num_rounds_new']][:10]
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

**Какие зарплаты предлагали продавцы:**

```python
values_1 = data[(data['player.id_in_group'] == 1) & (data['session.code'] == 'm2s8g1dl')].\
    groupby('participant.label')['group.salary'].apply(lambda x: list(x.values))
ends = [1, 5, 4, 9, 1]
for i, (index, val) in enumerate(values_1.iteritems()):
    values_1[index] = val[:ends[i]]
values_1
```

**Какие усилия прилагали работники:**

```python
values_2 = data[(data['player.id_in_group'] == 2) & (data['session.code'] == 'm2s8g1dl')].\
    groupby('participant.label')['group.staff'].apply(lambda x: list(x.values)[:9])
ends = [5, 4, 1, 1, 9]
for i, (index, val) in enumerate(values_2.iteritems()):
    values_2[index] = val[:ends[i]]
values_2
```

**Когда срывалась работа:**

```python
values_3 = data[(data['player.id_in_group'] == 2) & (data['session.code'] == 'm2s8g1dl')].\
    groupby('participant.label')['group.fail'].apply(lambda x: list(x.values)[:9])
for i, (index, val) in enumerate(values_3.iteritems()):
    values_3[index] = val[:ends[i]]
values_3
```

**В первой сессии все предложения фирмы были > 40, практически всегда работниками прикладывались усилия. Работа срывалась 3 раза у одной пары участников, во все случаи, когда участник не прикладывал усилия при зарплате в 45 и 50.**
